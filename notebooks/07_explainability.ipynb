# Model Explainability Walkthrough
Visualizing model focus areas using Grad-CAM and Vanilla Saliency Maps.

```python
import os
import sys
import matplotlib.pyplot as plt
from pathlib import Path

sys.path.append(os.path.abspath('..'))
```

## 1. Grad-CAM & Saliency Map Comparison Grid

```python
explain_dir = Path('../reports/figures/explainability')
classes = ['glioma', 'meningioma', 'notumor', 'pituitary']

for cls in classes:
    fig_path = explain_dir / f'explain_comparison_{cls}.png'
    if fig_path.exists():
        img = plt.imread(str(fig_path))
        plt.figure(figsize=(15, 5))
        plt.imshow(img)
        plt.title(f'Explainability Breakdown for {cls.upper()}', fontsize=14)
        plt.axis('off')
        plt.show()
```
